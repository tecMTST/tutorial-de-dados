```python
import pandas as pd
```

```python
df_receita = pd.read_csv('dados/processados/SP/2022/DeputadoFederal/doador_candidato.csv', encoding='latin-1', sep=',')
df_candidato = pd.read_csv('dados/processados/SP/2022/DeputadoFederal/candidato.csv', encoding='latin-1', sep=',')
df_rec_cand = pd.read_csv('dados/processados/SP/2022/DeputadoFederal/receita_candidato.csv', encoding='latin-1', sep=',')
```

```python
df_rec_cand[['DS_ORIGEM_RECEITA']].drop_duplicates()
```

```python
df_rec_cand = df_rec_cand[df_rec_cand['DS_ORIGEM_RECEITA'].isin(['Recursos de partido político', 'Recursos de pessoas físicas'])]

df_rec_part = pd.merge(df_candidato[['SG_PARTIDO', 'SQ_CANDIDATO']], df_rec_cand, on='SQ_CANDIDATO').pivot_table(index='SG_PARTIDO', columns='DS_ORIGEM_RECEITA', values='VR_RECEITA', aggfunc='sum', fill_value=0)

df_rec_part = df_rec_part.reset_index()

df_rec_part.columns = ['SG_PARTIDO', 'VR_RECEITA_PARTIDO', 'VR_RECEITA_PF']
```

```python
df_rec_part['TOT_RECEITA'] = df_rec_part['VR_RECEITA_PARTIDO'] + df_rec_part['VR_RECEITA_PF']
df_rec_part['PCT_RECEITA'] = df_rec_part['TOT_RECEITA']/df_rec_part['VR_RECEITA_PF']

df_rec_part.sort_values('DIF_RECEITA', ascending=False)
```

```python
df_rec_part = df_rec_cand[df_rec_cand['DS_ORIGEM_RECEITA']=='Recursos de partido político']

df_rec_cand = pd.merge(df_candidato[['SQ_CANDIDATO', 'NM_CANDIDATO', 'SG_PARTIDO', 'DS_SIT_TOT_TURNO']], df_rec_cand, on='SQ_CANDIDATO')

df_rec_part = df_rec_cand.groupby('SG_PARTIDO').agg({'VR_RECEITA': 'sum'}).reset_index()

df_rec_part.sort_values(['VR_RECEITA'], ascending=False)
```

```python
df_receita = pd.merge(df_candidato[['SQ_CANDIDATO', 'DS_SIT_TOT_TURNO']], df_receita, on='SQ_CANDIDATO')
```
